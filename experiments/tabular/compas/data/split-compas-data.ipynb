```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
#load full data
data = pd.read_csv('compas-scores.csv')  

#quick look
print(data.shape)
display(data.head())
display(data.describe())
```

```python
#remove data points with negative values for 'length-of-stay'
print('number of points where length-of-stay is -1:', sum(data['length-of-stay'] == -1))
data = data[data['length-of-stay'] != -1]
```

```python
#split train-test
train, test = train_test_split(data, test_size=0.20, random_state=12345)

#save splits
train.to_csv('compas-train.csv', index=False)
test.to_csv('compas-test.csv', index=False)
```

```python
print('data, shape:', data.shape)
print('train, shape:', train.shape)
print('test, shape:', test.shape)
```
